```python
import pandas as pd
import numpy as np
%matplotlib inline
from pathlib import Path
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
#loading the file
NetExport_df = pd.read_csv('C:/Users/User/Desktop/Second Degree/BA Honours in Economics/University of Calgary/2024 Winter Semester/Econ 588/Open Government - Natural Gas - Imports and Exports/natural-gas-exports-and-imports-monthly.csv')
```

```python
# Scatter plot of Exchange Rate (CADUSD) vs. Net Export (10^3 m3)mm
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Exchange Rate (CADUSD)', y='Net Export (10^3 m3)mm', data=NetExport_df)
plt.title('Exchange Rate (CADUSD) vs. Net Export (10^3 m3)mm')
plt.xlabel('Exchange Rate (CADUSD)')
plt.ylabel('Net Export (10^3 m3)mm')
plt.show()
```

```python
y = NetExport_df['Net Export (10^3 m3)mm']

X = NetExport_df['Exchange Rate (CADUSD)']
X = sm.add_constant(X)

results = sm.OLS(y,X).fit()
print(results.summary())

predictions = results.predict(X)

plt.plot(NetExport_df['Exchange Rate (CADUSD)'], predictions)
plt.scatter(NetExport_df['Exchange Rate (CADUSD)'], y)
plt.show()
```
